```python
def kelvin_to_celsius(k):
    return k - 273.15

def celsius_to_kelvin(c):
    return c + 273.15
```

```python
def to_fahrenheit(temp, scale="celsius"):
    if scale == "celsius":
        return (temp * 9/5) + 32
    elif scale == "kelvin":
        c = kelvin_to_celsius(temp)
        return (c * 9/5) + 32
```

```python
print(to_fahrenheit(0))
print(to_fahrenheit(273.15, "kelvin"))
```

```python
def from_fahrenheit(f, output="celsius"):
    c = (f - 32) * 5/9
    
    if output == "celsius":
        return c
    elif output == "kelvin":
        return celsius_to_kelvin(c)
```

```python
print(from_fahrenheit(32))
print(from_fahrenheit(32, "kelvin"))
```

```python
def height_to_meters(feet, inches):
    return feet * 0.3048 + inches * 0.0254
```

```python
def meters_to_height(meters):
    total_inches = meters / 0.0254
    feet = int(total_inches // 12)
    inches = total_inches % 12
    return feet, inches
```

```python
heights = [1.5, 1.7, 1.85]

for h in heights:
    feet, inches = meters_to_height(h)
    converted_back = height_to_meters(feet, inches)
    print("Original:", h)
    print("Round-trip:", converted_back)
    print()
```
